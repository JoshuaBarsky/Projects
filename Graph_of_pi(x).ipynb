Graphing the prime counting function.

```python
import mpmath as mp
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def primesuptolist(n):
  flags = np.full(n,True)
  flags[:2] = False
  p = 2
  uptolist = np.zeros(n)
  count = 0
  while p < n:
    if flags[p] == True:
      flags[p*p::p] = [False] * len(flags[p*p::p])
      count += 1
      uptolist[p] = count
    else:
      uptolist[p] = count
    p += 1
  return uptolist
```

```python
xval = np.arange(0,100)
yval = primesuptolist(100)
plt.step(xval,yval, where='post')
plt.title('$\pi(x)$ for $0<x<100$')
plt.xlabel('$x$')
plt.ylabel('Number of primes')
plt.grid(True)
plt.show()
```
