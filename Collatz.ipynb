##**Collatz Conjecture**

This program maps the Collatz Conjecture.\
https://mathworld.wolfram.com/CollatzProblem.html

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def collatz(n):
  if n%2 == 0:
    n = n/2
    return [n] + collatz(n)
  elif n%2 != 0 and n != 1:
    n = 3*n+1
    return [n] + collatz(n)
  else:
    return [n]
```

```python
iterations = 125
```

```python
x_values = [i+1 for i in range(iterations)]
y_values = [[] for i in range(iterations)]

for i in range(len(x_values)):
  y_values[i] = collatz(i+1)

for i in range(len(x_values)):
  if len(y_values[i]) < iterations: #if the y values are too short, add zeros.
    for j in range(iterations - len(y_values[i])):
      y_values[i].append(0)
  if len(y_values[i]) > iterations: #if the y values are too long, delete last elements.
    for j in range(len(y_values[i]) - iterations):
      y_values[i].pop()
```

```python
plt.figure(figsize=(20, 10));
plt.xlabel("iterations")
plt.ylabel("output")

for i in range(len(x_values)):
  plt.plot(x_values, y_values[i]);
```
